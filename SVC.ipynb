```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,:len(df.columns)-1],df.iloc[:,len(df.columns)-1].astype(int),shuffle=True,random_state=0)
x_train.shape
```

```python
model=SVC(C=1,kernel='linear')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

# *Get Best Parameter*

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
param_grid = {
    'C': [0.1, 1, 10, 100], 
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'degree': [2, 3, 4],  # used for 'poly' kernel
    'gamma': ['scale', 'auto']  # used for 'rbf', 'poly', and 'sigmoid' kernels
}

# Initialize the model
svc = SVC()

# Perform grid search
grid_search = GridSearchCV(estimator=svc, param_grid=param_grid, cv=5, scoring='accuracy')

# Fit the grid search
grid_search.fit(x_train, y_train)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Parameters:", best_params)
print("Best Cross-Validation Score:", best_score)

# Train the final model with the best parameters
best_model = grid_search.best_estimator_
best_model.fit(x_train, y_train)

# Evaluate the model on the test set
y_pred = best_model.predict(x_test)
test_accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy:", test_accuracy)
```
